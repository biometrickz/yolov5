```python
df1 = pd.read_excel('rus_xlsx/my_file.xlsx', index_col=None)
df2 = pd.read_excel('rus_xlsx/my_file_2_kundyz.xlsx', index_col=None)
df3 = pd.read_excel('rus_xlsx/my_file3.xlsx', index_col=None)
df4 = pd.read_excel('rus_xlsx/my_file_kundyz.xlsx', index_col=None)
df5 = pd.read_excel('rus_xlsx/Batch2_Aslan.xlsx', index_col=None)
df6 = pd.read_excel('rus_xlsx/my_file2.xlsx', index_col=None)
df7 = pd.read_excel('rus_xlsx/batch2.xlsx', index_col=None)
df8 = pd.read_excel('rus_xlsx/Batch1_Aslan.xlsx', index_col=None)
```

```python
df_List = [df1,df2,df3,df5,df4,df6,df7,df8]
concat_df = pd.concat(df_List)
concat_df = concat_df.loc[:, ~concat_df.columns.str.contains('^Unnamed')]
# concat_df = concat_df.dropna()

concat_df = concat_df.reset_index(drop = True)

concat_df.shape
```

```python
concat_df.drop(concat_df[concat_df.side == 'NaN'].index, inplace=True)
```

```python
c = 0
for index, row in concat_df.iterrows():
    if type(row['filename'])== str:
        if os.path.exists(os.path.join('/home/user/ml_projects/Myrzakhan/ru_annotation/all_batches_first_layer/',row['filename'])):
            # print(row['filename'])
            img = cv2.imread(os.path.join('/home/user/ml_projects/Myrzakhan/ru_annotation/all_batches_first_layer/',row['filename']))
            if img is not None:
                height, width, _  = img.shape
                img_name = row['filename'].split('.')[0]
                save_txt_path = os.path.join('/home/user/ml_projects/Myrzakhan/ru_annotation/all_batches_first_layer',row['filename']) + '.txt'
                if row['side']=='B' or row['side']=='b':
                    str_label = '2 '
                    with open(save_txt_path, "w") as f:
                        # print(row['x1'],row['x2'],row['x3'],row['x4'],row['x5'],row['x6'])
                        x_y_coord =min(row['x1'],row['x4']),min(row['y2'],row['y1'])
                        # print('x_y_coord',x_y_coord)
                        w_h_coord = max(row['x3'],row['x2']),max(row['y3'],row['y4'])
                        str_label += str((x_y_coord[0] + (w_h_coord[0]-x_y_coord[0])/2)/width) +' '
                        str_label += str((x_y_coord[1]+(w_h_coord[1]-x_y_coord[1])/2)/height) +' '
                        str_label += str((w_h_coord[0]-x_y_coord[0])/width) +' '
                        str_label += str((w_h_coord[1]-x_y_coord[1])/height) + '\n'
                        f.write(str_label)
                elif row['side']=='F' or row['side']=='f':
                    str_label = '1 '
                    with open(save_txt_path, "w") as f:
                        # print(row['x1'],row['x2'],row['x3'],row['x4'],row['x5'],row['x6'])
                        x_y_coord =min(row['x1'],row['x4']),min(row['y2'],row['y1'])
                        # print('x_y_coord',x_y_coord)
                        w_h_coord = max(row['x3'],row['x2']),max(row['y3'],row['y4'])
                        str_label += str((x_y_coord[0] + (w_h_coord[0]-x_y_coord[0])/2)/width) +' '
                        str_label += str((x_y_coord[1]+(w_h_coord[1]-x_y_coord[1])/2)/height) +' '
                        str_label += str((w_h_coord[0]-x_y_coord[0])/width) +' '
                        str_label += str((w_h_coord[1]-x_y_coord[1])/height) + '\n'
                        f.write(str_label)
                else:
                    str_label = '0 '
                    with open(save_txt_path, "w") as f:
                        # print(row['x1'],row['x2'],row['x3'],row['x4'],row['x5'],row['x6'])
                        x_y_coord =min(row['x1'],row['x5']),min(row['y2'],row['y1'])
                        # print('x_y_coord',x_y_coord)
                        w_h_coord = max(row['x4'],row['x2']),max(row['y4'],row['y5'])
                        str_label += str((x_y_coord[0] + (w_h_coord[0]-x_y_coord[0])/2)/width) +' '
                        str_label += str((x_y_coord[1]+(w_h_coord[1]-x_y_coord[1])/2)/height) +' '
                        str_label += str((w_h_coord[0]-x_y_coord[0])/width) +' '
                        str_label += str((w_h_coord[1]-x_y_coord[1])/height) + '\n'
                        f.write(str_label)

# print(c)
```
