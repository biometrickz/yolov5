```python

# Import pandas
import pandas as pd
import numpy as np
  
# reading csv file 
df = pd.read_csv("3090_этап_1_v2.csv")
df = df.replace(np.nan, 111)
```

```python
import os
import cv2
```

```python
df.head()
```

```python
for index, row in df.iterrows():
    img = cv2.imread(os.path.join('images1',row['filenames']))
    h, w, c = img.shape
    # font
    font = cv2.FONT_HERSHEY_SIMPLEX
    
    # org
    org = (int(w/2)-250, int(h/2))
    
    # fontScale
    fontScale = 1
    
    # Blue color in BGR
    color = (255, 0, 0)
    
    # Line thickness of 2 px
    thickness = 2
   
    image = cv2.putText(img, 'side'+str(row['side']) + '>'+'ver'+str(row['document_version']) +'>'+'rot_deg'+str(row['rotation_degree']), org, font, 
                   fontScale, color, thickness, cv2.LINE_AA)
    

    center_coordinates=(int(row['x1']),int(row['y1']))
    radius=5
    color=(0,0,255)
    thickness=-1

    image = cv2.circle(image, center_coordinates, radius, color, thickness)
    center_coordinates=(int(row['x2']),int(row['y2']))
    color=(0,255,0)
    image = cv2.circle(image, center_coordinates, radius, color, thickness)
    center_coordinates=(int(row['x3']),int(row['y3']))
    color=(255,0,0)
    image = cv2.circle(image, center_coordinates, radius, color, thickness)
    center_coordinates=(int(row['x4']),int(row['y4']))
    color=(0,0,0)
    image = cv2.circle(image, center_coordinates, radius, color, thickness)
    cv2.imwrite(os.path.join('images1_out',row['filenames']),image)
```
